```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


from place_cells import PlaceCells
```

```python
#Place cells options
class Options:
    pass
options = Options()

options.Np = 512
options.place_cell_rf = 12 # width of place cell center tuning curve (m)
options.surround_scale = 2 # if DoG, ratio of sigma2^2 to sigma1^2
options.box_width = 2.2
options.box_height = 2.2
options.DoG = True
options.periodic = False
options.topology='klein'


#generate place cells
place_cells = PlaceCells(options)
```

```python
#Plot and save place cells centers
us = place_cells.us
plt.figure(figsize=(6,6))
plt.scatter(us[:,0], us[:,1],s=4, label='Place cell centers')
plt.title('Place cell centers')
```

```python
#Plot with arena border
x = place_cells.us[:,0]
y = place_cells.us[:,1]
fig = plt.figure(figsize=(6,6))
axes= fig.add_axes([0.1,0.1,0.9,0.9])
axes.scatter(x,y,s=4)

#plot arena
h0 = -options.box_height/2
h1 = options.box_height
w0 = -options.box_width/2
w1 = options.box_width
arena = Rectangle((w0,h0),w1,h1, fill=False)


axes.add_patch(arena)
axes.set_title('Place cells centers')
axes.set_xlabel('x')
axes.set_ylabel('y');
```
