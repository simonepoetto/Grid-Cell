```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import ndimage
import PIL

from persim import plot_diagrams
from ripser import ripser, lower_star_img
```

```python
#Load activations
activations = np.load('../data/activations4096.npy')


#Load from png
#image = Image.open('image.png')
#image = image.convert('L') #convert to greyscale
#matrixi = np.asarray(image) #convert to matrix
```

```python
idx = 0
dgm = lower_star_img(-activations[idx])

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(activations[idx])
plt.colorbar()
plt.title("Grid Cell")
plt.subplot(122)
plot_diagrams(dgm)
plt.title("0-D Persistence Diagram")
plt.tight_layout()
plt.show()
```

```python
n_plot = 10
step = 200
plt.figure(figsize=(10,12))
for i in range(n_plot):
    plt.subplot(n_plot//2,4,2*i+1)
    plt.imshow(activations[i*step+110])
    plt.axis('off')
    plt.subplot(n_plot//2,4,2*i+2)
    dgm = lower_star_img(-activations[i*step+110])
    plot_diagrams(dgm)
    plt.tight_layout()
plt.show()
```
